```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## birinci kısım matplotlib

```python
x = np.arange(6) # burada x lerimizi 0 dan 6 ya kadar olan değerlerimiz
y = np.sin(x) # burada y değerlerimiz x değerlerin sin() değerleridir
```

```python
plt.plot(x,y,color = "green") # burada çizgi grafiğimizi oluşturuyoruz
plt.xlabel("x değerleri") # x eksenimize isimlendirme yapıyoruz
plt.ylabel("y değerleri") # y eksenimize isimlendirme yapıyoruz
plt.show()
```

```python
meyveler = {"elma" : 10 , "armut" : 15 , "muz" : 7 , "çilek" : 12} # verilerimizi oluşturduk
```

```python
plt.bar(meyveler.keys() , meyveler.values(), color ="blue") # burada ise çubuk grefiğimiz oluşturyoruz
plt.xlabel("meyve")
plt.ylabel("adet")
plt.title("meyve-adet-tablosu")
plt.show()
```

```python

# Rastgele veri oluşturma
x = np.random.rand(50) * 10  # 0 ile 10 arasında 50 rastgele sayı
y = np.random.rand(50) * 100 # 0 ile 100 arasında 50 rastgele sayı
data = np.random.randn(1000)  # 1000 adet rastgele veri noktası

# Yeni bir figür oluştur
plt.figure(figsize=(12, 5))  # Geniş bir figür ayarla

# 1. Grafik: Scatter Plot
plt.subplot(1, 2, 1)  # 1 satır, 2 sütun, 1. grafik
plt.scatter(x, y, color="purple", marker="o")
plt.xlabel("x değerleri")
plt.ylabel("y değerleri")
plt.title("Scatter Plot")

# 2. Grafik: Histogram
plt.subplot(1, 2, 2)  # 1 satır, 2 sütun, 2. grafik
plt.hist(data, bins=20, color="orange")
plt.xlabel("Değerler")
plt.ylabel("Frekans")
plt.title("Matplotlib ile Histogram")

# Tüm grafikleri göster
plt.tight_layout()  # Grafiklerin üst üste binmesini önler
plt.show()
```

## ikinci kısım seaborn


```python
data = {
    "İsim": ["Ahmet", "Mehmet", "Ayşe", "Fatma", "Ali"],
    "Yaş": [25, 30, 22, 28, 35],
    "Maaş": [5000, 7000, 4500, 6000, 8000],
    "Şehir": ["İstanbul", "Ankara", "İzmir", "Bursa", "Antalya"]
}

# DataFrame oluşturma
df = pd.DataFrame(data)
```

```python
sns.barplot(data = df,x = df["İsim"],y= df["Maaş"],color="cyan") # burada seaborn kullanarak df den çekilen verilerle bir çubuk grafiği çizilmiştir
plt.title("kişişel-gelir")
plt.show()
```

```python
sns.boxplot(data = df,x = df["İsim"],y= df["Maaş"],color="cyan") # burada seaborn kullanarak df den çekilen verilerle bir kutu grafiği çizilmiştir
plt.title("kişişel-gelir")
plt.show()
```
